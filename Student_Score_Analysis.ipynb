```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Student_Data_Input.csv")
print(df.head())
```

# Drop unnamed column

```python
df=df.drop("Unnamed: 0",axis=1)
```

```python
print(df.head())
```

```python
df.head()
```

# Gender distribution

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

# From the above chart, the no of FEMALE is more than the no of MALE

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb,annot=True)
plt.title("Relationship between Parental Education and Student's Score")
plt.show()
```

# From the above chart, it is seen that that the parental education has high impact on the scores of the students

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1,annot=True)
plt.title("Relationship between Parental Marital Status and Student's Score")
plt.show()
```

# From the above chart, we see that parental marital status has negligible impact on the marks of the students |

# Distribution of ETHNIC GROUPS -------->

```python
GroupA=df.loc[(df["EthnicGroup"]=="group A")].count()
GroupB=df.loc[(df["EthnicGroup"]=="group B")].count()
GroupC=df.loc[(df["EthnicGroup"]=="group C")].count()
GroupD=df.loc[(df["EthnicGroup"]=="group D")].count()
GroupE=df.loc[(df["EthnicGroup"]=="group E")].count()
l=['group A','group B','group C','group D','group E']
lists=[GroupA['EthnicGroup'],GroupB['EthnicGroup'],GroupC['EthnicGroup'],GroupD['EthnicGroup'],GroupE['EthnicGroup']]
plt.pie(lists,labels=l,autopct="%1.2f%%")
plt.title("Distribution of ETHNIC GROUPS")
plt.show()
```
